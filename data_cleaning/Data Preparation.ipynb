##   Step 1: Build the dataframe
Create a pandas dataframe for data learning, and future exploratory data analysis (EDA) and statistical activities

```python
import pandas as pd               #library exercise for buidling dataframes
import numpy as np                #numpy is imported with pandas
#Load data from the CSV extracted from the database
df1=pd.read_csv("data/requisition1.csv")
df2=pd.read_csv("data/requisition2.csv")
df3=pd.read_csv("data/requisition3.csv")
df4=pd.read_csv("data/requisition4.csv")
df5=pd.read_csv("data/requisition5.csv")
df6=pd.read_csv("data/requisition6.csv")
```

```python
# Data translation (To skip or ignor) during the post of scripts
fr_to_en={'Août':'Aug'}
for fr,en in fr_to_en.items():
    df1['processing_periods_name']=df1['processing_periods_name'].str.replace(fr,en,regex=False)
    df2['processing_periods_name']=df2['processing_periods_name'].str.replace(fr,en,regex=False)
    df3['processing_periods_name']=df3['processing_periods_name'].str.replace(fr,en,regex=False)
    df4['processing_periods_name']=df4['processing_periods_name'].str.replace(fr,en,regex=False)
    df5['processing_periods_name']=df5['processing_periods_name'].str.replace(fr,en,regex=False)
    df6['processing_periods_name']=df6['processing_periods_name'].str.replace(fr,en,regex=False)
```

```python
df1.to_csv("data/requisition1.csv",index=False)
df2.to_csv("data/requisition2.csv",index=False)
df3.to_csv("data/requisition3.csv",index=False)
df4.to_csv("data/requisition4.csv",index=False)
df5.to_csv("data/requisition5.csv",index=False)
df6.to_csv("data/requisition6.csv",index=False)
```

#### 2.2. Dealing with missing values

```python
requisitions_all.shape[0]
```

```python
requisitions_all.isna().sum()
```

Let's represent the missing values in %

```python
(requisitions_all.isna().sum()/requisitions_all.shape[0])*100
```

* There are missing values in the following columns:beginningbalance => 48956, quantityreceived => 48696. etc. Same for quantitydispensed,stockinhand,quantityrequested, quantityapproved,quantityapproved,amc.
* The core inventory variables beginningbalance,quantityreceived,quantitydispensed,stockinhand,amc are all representing ~ 28% (48,000) observations.These variables are jointly missing, which is very important could suggest entire reports are missing (not individual fields). This may be due to the reporting completeness issue, facilities that did NOT submit reports for that month/product as the product is not available, not used.
* quantityrequested represent ~72% missing(123604), this pattern could be justified as requests only happen in order's months - quarterly (requests only happen in order months) so Feb, May, Aug, Nov with no requests and 4 months with requests.
* Quantityapproved represent ~19% missing(31989). This is lower than quantity requested, it could du to the fact that the orders are placed even for facilities with no requested quantities (pushing strategy formalized)

```python
# Display the missing data 
mask=requisitions_all[
    ["beginningbalance","quantityreceived","quantitydispensed","stockinhand"]
].isna().all(axis=1)
mask_amc=requisitions_all[
    ["amc"]
].isna().all(axis=1)
```

```python
# Remove missing data
requisitions_cleaned_inventory=requisitions_all.dropna(
    subset=[
        "beginningbalance",
        "quantityreceived",
        "quantitydispensed",
        "stockinhand"
    ]
)
```
